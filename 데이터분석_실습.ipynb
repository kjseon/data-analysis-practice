# Lecture 10_2 선형회귀분석 (실습)
연습문제 1 : 키 - 몸무게 데이터

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# (1) 데이터 준비
data = [(174, 71), (152, 55), (160, 46), (137, 40), (180, 85)]
df = pd.DataFrame(data, columns=['height', 'weight'])

# (2) X, y 분리
X = df[['height']]
y = df['weight']

# (3) 선형 회귀 모델 학습
model = LinearRegression()
model.fit(X, y)

# (4) 예측
height_to_predict = 165
predicted_weight = model.predict([[height_to_predict]])[0]

# (5) 모델 파라미터 및 정확도 출력
slope = model.coef_[0]
intercept = model.intercept_
score = model.score(X, y)
r2 = r2_score(y, model.predict(X))
print(f"기울기(Weight): {slope:.4f}")
print(f"절편(Intercept): {intercept:.4f}")
print(score)
print(f"R²(예측 정확도): {r2:.4f}")
print(f"키가 {height_to_predict}cm일 때 예측 몸무게: {predicted_weight:.2f}kg")

# (6) 회귀선 시각화
plt.scatter(df['height'], df['weight'], color='blue', label='Actual Data')
plt.plot(df['height'], model.predict(X), color='red', label='Regression Line')
plt.scatter(height_to_predict, predicted_weight, color='green', s=100, label=f'Predicted (165cm)')
plt.title('Linear Regression: Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend()
plt.grid(True)
plt.show()
```

연습문제 2 : 주당 운동 시간

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# (1) 데이터 준비
data = {
    'exercise_hour_per_week': [2, 4, 6, 8, 10, 12],
    'weight': [68.4, 67.3, 65.4, 65.2, 64.9, 64.6]
}
df = pd.DataFrame(data)

# (2) X, y 분리
X = df[['exercise_hour_per_week']]
y = df['weight']

# (3) 선형 회귀 모델 학습
model = LinearRegression()
model.fit(X, y)

# (4) 예측
predict_hour = 5
predicted_weight = model.predict([[predict_hour]])[0]

# (5) 결과 출력
slope = model.coef_[0]
intercept = model.intercept_
r2 = r2_score(y, model.predict(X))

print(f"기울기(Weight): {slope:.4f}")
print(f"절편(Intercept): {intercept:.4f}")
print(f"R²(예측 정확도): {r2:.4f}")
print(f"운동 {predict_hour}시간/주일 때 예측 몸무게: {predicted_weight:.2f}kg")

# (6) 시각화
plt.scatter(df['exercise_hour_per_week'], df['weight'], color='blue', label='Actual Data')
plt.plot(df['exercise_hour_per_week'], model.predict(X), color='red', label='Regression Line')
plt.scatter(predict_hour, predicted_weight, color='green', s=100, label=f'Predicted (5h)')
plt.xlabel('time')
plt.ylabel('weight (kg)')
plt.legend()
plt.grid(True)
plt.show()
```

# Lecture 11_2 클러스터링(군집화)
실습 1: 쇼핑몰 고객분류

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler # StandardScaler import 추가
from yellowbrick.cluster import KElbowVisualizer # KElbowVisualizer import 추가

# 0. 데이터 불러오기
mall = pd.read_csv('/content/sample_data/Mall_Customers.csv')
mall.info()
mall.head(5)

#1. 탐색적 데이터 분석
# 구매점수 분포
plt.figure(1,figsize=(15,6))
sns.histplot(mall['Spending Score (1-100)'], bins = 20, kde=True) # distplot을 histplot으로 변경
plt.show()
# 나이별 분포
plt.figure(1,figsize=(15,6))
sns.displot(mall['Age'], bins = 20) # displot 사용
plt.show()
# 연간 소득 분포
plt.figure(1,figsize=(15,6))
sns.histplot(mall['Annual Income (k$)'], bins = 20, kde=True) # distplot을 histplot으로 변경
plt.show()
# 전체 데이터 분포
sns.pairplot(mall[['Age','Annual Income (k$)', 'Spending Score (1-100)']])
plt.suptitle('Feature Pair Relationships', y=1.02)
plt.show()

#2.target / source 설정
target_col = 'Spending Score (1-100)'      # 타깃 (우리가 설명하고 싶은 값)
source_cols = ['Annual Income (k$)', 'Age']  # 각각의 원인 후보 변수들

#3. KMEANS 반복 실험
for src in source_cols:
    print(f"\n==============================")
    print(f"Target: {target_col} | Source: {src}")
    print("==============================")

    # 2D 데이터 구성 (target + source)
    X = mall[[target_col, src]]

    # 데이터 스케일링 (KMeans와 Silhouette Score에 중요)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # k=5로 KMeans 수행
    print(f"\n--- KMeans Clustering (k=5) for {src} vs {target_col} ---")
    kmeans = KMeans(n_clusters=5, random_state=42, n_init='auto')
    mall['Cluster'] = kmeans.fit_predict(X_scaled) # 스케일링된 데이터로 fit_predict

    # k=5 시각화
    plt.figure(figsize=(7,5))
    sns.scatterplot(data=mall, x=src, y=target_col, hue='Cluster', palette='tab10', s=40)
    plt.title(f'KMeans Clustering: {src} vs {target_col}')
    plt.xlabel(src)
    plt.ylabel(target_col)
    plt.legend(title='Cluster')
    plt.show()

    # k=5 군집별 평균 출력
    print(mall.groupby('Cluster')[[src, target_col]].mean().round(2))

    print(f"\n--- Optimal k KMeans Clustering for {src} vs {target_col} ---")
    # KElbowVisualizer를 사용하여 최적의 k 값 찾기
    model_elbow = KMeans(random_state=42, n_init='auto')
    visualizer = KElbowVisualizer(model_elbow, k=(2,10), metric='silhouette', timings=False, title='Silhouette score for Elbow for KMeans clustering')
    visualizer.fit(X_scaled)
    visualizer.show()

    best_k = visualizer.elbow_value_
    print(f"Optimal k for {src} vs {target_col}: {best_k}")

    # 군집별 평균 출력
    print(mall.groupby('Cluster')[[src, target_col]].mean().round(2))
```

실습 2: 와인 데이터 분류

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

# 0. 데이터 불러오기
wine = pd.read_csv('/content/sample_data/wine-clustering.csv')
wine.info()
wine.head(5)
#1. 탐색적 데이터 분석
sns.pairplot(wine)
plt.suptitle('Pairplot of Wine Dataset', y=1.02)
plt.show()

# (1) < Alcohol vs Color Intensity >
#2. 데이터 구성
X_1 = wine[['Alcohol', 'Color_Intensity']]
X_1 = wine.select_dtypes(include = ['float64', 'int64']) # 수치형 feature만 사용
scaler1 = StandardScaler()
X_scaled_1 = scaler1.fit_transform(X_1)

#3. 클러스터링 : KMeans + KElbowVisualizer
# KMeans 모델 생성 및 KElbow로 최적 k 찾기
model1 = KMeans(random_state=42, n_init='auto')
visualizer1 = KElbowVisualizer(model1, k=(2,10),metric='silhouette', timings=False)
visualizer1.fit(X_scaled_1)
visualizer1.show()

best_k1 = visualizer1.elbow_value_
print(f"최적의 k 값(Alcohol vs Color Intensity): {best_k1}")
# 4. 최적 k로 클러스터링 수행
kmeans1 = KMeans(n_clusters=best_k1, random_state=42, n_init='auto')
wine['Cluster(1)'] = kmeans1.fit_predict(X_scaled_1)

# 5. 시각화
plt.figure(figsize=(10, 7))
sns.scatterplot(data=wine, x='Alcohol', y='Color_Intensity', hue='Cluster(1)', palette='viridis', s=100, alpha=0.8)
plt.title('K-Means Clustering of Wine: Alcohol vs Color Intensity')
plt.xlabel('Alcohol')
plt.ylabel('Color_Intensity')
plt.legend(title='Cluster(1)')
plt.grid(True)
plt.show()

# 6. 군집별 평균 출력
print("\nMean values per cluster (Alcohol vs Color Intensity):")
print(wine.groupby('Cluster(1)')[['Alcohol', 'Color_Intensity']].mean().round(2))

# (2) < Alcohol vs Flavanoids >
#2. 데이터 구성
X_2 = wine[['Alcohol', 'Flavanoids']]
X_2 = wine.select_dtypes(include = ['float64', 'int64']) # 수치형 feature만 사용
scaler2 = StandardScaler()
X_scaled_2 = scaler2.fit_transform(X_2)

#3. 클러스터링 : KMeans + KElbowVisualizer
# KMeans 모델 생성 및 KElbow로 최적 k 찾기
model2 = KMeans(random_state=42, n_init='auto')
visualizer2 = KElbowVisualizer(model2, k=(2, 10), metric='silhouette', timings=False)
visualizer2.fit(X_scaled_2)
visualizer2.show()

best_k2 = visualizer2.elbow_value_
print(f"최적의 k 값(Alcohol and Flavanoids): {best_k2}")

# 4. 최적 k로 클러스터링 수행
kmeans2 = KMeans(n_clusters=best_k2, random_state=42, n_init='auto')
wine['Cluster(2)'] = kmeans2.fit_predict(X_scaled_2)

# 5. 시각화
plt.figure(figsize=(10, 7))
sns.scatterplot(data=wine, x='Alcohol', y='Flavanoids', hue='Cluster(2)', palette='viridis', s=100, alpha=0.8)
plt.title('K-Means Clustering of Wine: Alcohol vs Flavanoids')
plt.xlabel('Alcohol')
plt.ylabel('Flavanoids')
plt.legend(title='Cluster(2)')
plt.grid(True)
plt.show()

# 6. 군집별 평균 출력
print("\nMean values per cluster (Alcohol vs Flavanoids):")
print(wine.groupby('Cluster(2)')[['Alcohol', 'Flavanoids']].mean().round(2))
```

# Lecture 12_1 Classification(분류)_KNN

실습: 당뇨병 예측 여부

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 0. 데이터 불러오기
dia = pd.read_csv('/content/sample_data/diabetes.csv')
dia.info()
dia.head(5)

# 1. X,y 나누기 (Outcome 제외 = X, Outcome = y)
X = dia.drop('Outcome', axis=1)   # 입력 특성들
y = dia['Outcome']    # 정답(레이블)

# 2. train_test_split으로 8:2로 나누기
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, # 20%를 시험용으로
                                                    random_state=42,stratify=y) # 시드 고정, 클래스 비율 유지

# 3. StandardScaler()로 정규화 (훈련 데이터 기준으로만 fit)
scaler = StandardScaler()
scaler.fit(X_train)         # 훈련 데이터로만 평균/표준편차 계산
X_train_scaled = scaler.transform(X_train) # fit_transform 대신 transform 사용 (fit이 이미 호출되었으므로)
X_test_scaled = scaler.transform(X_test)

# 4. k에 따른 최적의 k 값 탐색 (여기서는 1~20까지)
best_k = 1
best_score = 0.0
print("\n=== k에 따른 정확도(시험 데이터 기준) ===")
for k in range(1, 11):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    y_pred = knn.predict(X_test_scaled)
    score = knn.score(X_test_scaled, y_test)
    print(f"k={k:2d}, accuracy={score:.3f}")
    # 가장 높은 정확도일 때 best_k 갱신
    if score > best_score:
        best_score = score
        best_k = k
# 최적의 k 값 출력은 for 루프 밖으로 이동
print(f"\n최적의 k 값: {best_k}, 해당 정확도: {best_score:.3f}")

# 5. 최적의k 값으로최종모델학습+ 예측결과확인
print("\n=== 최종모델결과(앞20개) ===")
final_knn = KNeighborsClassifier(n_neighbors=best_k)
final_knn.fit(X_train_scaled, y_train)
final_pred = final_knn.predict(X_test_scaled)
result_df = pd.DataFrame({
    '실제값': y_test.values,
    '예측값': final_pred
})
print(result_df.head(20))
print(f"\n최종정확도: {accuracy_score(y_test, final_pred):.3f}")
```

# Lecture 12_2 Classification(분류)_Logistic Regression

실습: 당뇨병 예측 여부

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# 0. 데이터 불러오기
dia = pd.read_csv('/content/sample_data/diabetes.csv')
print(dia.head())
print(dia.info())

# 1. X, y 나누기 (Outcome 제외 = X, Outcome = y)
X = dia.drop('Outcome', axis=1)  # 입력 특성들
y = dia['Outcome']               # 정답(레이블)
# 2. train_test_split으로 8:2로 나누기
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, #20%를 시험용으로
random_state=42, stratify=y ) #시드 고정, 클래스 비율 유지

# 4. 표준화(스케일링)
schaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 5) 로지스틱 회귀 학습
model = LogisticRegression()
model.fit(X_train_scaled, y_train)

# 6) 예측
y_pred= model.predict(X_test_scaled)

# 7) 평가
acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print("\n=== Accuracy ===")
print(f"{acc:.3f}")
print("\n=== Confusion Matrix ===")
print(cm)
print("\n=== Classification Report ===")
print(classification_report(y_test, y_pred))
```
